```python
data = {"Name":["a","b","c","d"],"Age":[10,20,30,40],"Salary":[100,200,np.nan,400]}
```

```python
new_df2 = pd.DataFrame(data)
```

```python
new_df2
```

```python
# dealing with null values.
# .dropna() --> removes the null values.
# .fillna() --> replaces the null values with mean,mode,median,0, categorical data to 'Unknown.
# 1. using fillna with mean.
new_df2['Salary'].fillna(new_df2['Salary'].mean())
```

```python
# mean --> continuous data
# median -->discrete numeric or sorted data
# mode --> categorical data
# 2. ffill()
# 3. bfill()
# 4. dropna()
new_df2['Salary'].ffill()
```

```python
new_df2['Salary'].bfill()
```

```python
# Merging data. (by default --> inner join)
df1_data = {'Id':[0,1,2,3,4],'Name':['A','B','C','D','E'], 'Salary':[10,20,30,40,50]}
df2_data = {'Id':[2,3,1,0],'Dept':['Sales','Marketing','Finance','Marketing']}
```

```python
df1 = pd.DataFrame(df1_data)
df2 = pd.DataFrame(df2_data)
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,left_on='Id',right_on='Id',how = "outer")
```
