```python
# Data Transformation
# pd.melt() to change wide form data to long form data
d1 = {"Name": ["Pankaj", "Lisa", "David"], "ID": [1, 2, 3], "Role": ["CEO", "Editor", "Author"]}

df2 = pd.DataFrame(d1)

print(df2)

df_melted = pd.melt(df2, id_vars=["ID"], value_vars=["Name", "Role"], var_name="Attribute", value_name="Value")

print(df_melted)
```

```python
# unmelting using pivot()

df_unmelted = df_melted.pivot(index='ID', columns='Attribute')

print(df_unmelted)
```

```python
df_unmelted = df_unmelted['Value'].reset_index()
df_unmelted.columns.name = None
print(df_unmelted)
```

```python
# merge() function is used to horizontally join 2 dfs
# Syntax - pd.merge(df1, df2, how='', on='', right_on, left_on)
# types of merges -> inner, outer, left, right, cross
```

```python
data1 = {'Id':[1,2,3,4,5], 'Name':["A","B","C","D","E"], 'Salary':[100, 150, 230,250,180]}
data2 = {'Dept_id': [1,2,4,6,7], 'Dept_name': ['Sale','Marketing','Finance','Sales','Marketing']}
```

```python
df_a = pd.DataFrame(data1)
df_b = pd.DataFrame(data2)
```

```python
# default - inner join
df_a.merge(df_b, left_on='Id', right_on='Dept_id')
```

```python
# left join
df_a.merge(df_b, how='left',left_on='Id', right_on='Dept_id')
```

```python
# right join 
df_a.merge(df_b,how='right', left_on='Id', right_on='Dept_id')
```

```python
# outer join 
df_a.merge(df_b, how='outer',left_on='Id', right_on='Dept_id')
```

```python
# cross join
df_a.merge(df_b,how='cross')
```

```python
# df.align - to align indexes and columns of two dataframes

df_r = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=[0, 1, 2])
df_s = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]}, index=[1, 2, 3])

align1 , align2 = df_r.align(df_s)
print(align1)
print(align2)
```

```python
# Creating MultiIndex DataFrame
arrays = [['A', 'A', 'B', 'B'], ['1', '2', '1', '2']]
midx = pd.MultiIndex.from_arrays(arrays, names=('letter', 'number'))
df_multi = pd.DataFrame({'val': [10, 20, 30, 40]}, index=midx)
print(df_multi)
# Accessing with multiple levels
print(df_multi.loc['A'])        # All rows with first level 'A'
print(df_multi.loc[('A', '1')])  # Specific row
```

```python
# Window functions in Pandas - like running averages, totals, min and max
s_data = {'Value': [10, 12, 15, 11, 18, 13]}
df_running = pd.DataFrame(s_data)

# Rolling mean with a window of 3
df_running['Rolling Mean'] = df_running['Value'].rolling(window=3).mean()
print(df_running) # first two values are NaN because of 3 day average
```

```python
df_running['Running Sum'] = df_running['Value'].expanding().sum() # to get running sum total
df_running
```
